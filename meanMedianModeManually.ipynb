# Mean, Median, and Mode - Manually

## Task

Given an array, X, of N integers, calculate and print the respective mean, median, and mode on separate lines. If your array contains more than one modal value, choose the numerically smallest one.

**Note:** Other than the modal value (which will always be an integer), your answers should be in decimal form, rounded to a scale of 1 decimal place (i.e., 12.3, 7.0 format).

## Input Format

The first line contains an integer, N, denoting the number of elements in the array. 

The second line contains N space-separated integers describing the array's elements.

## Constraints

- The number of elements N shall be greater than or equal to 10, and less than or equal to 2500.
- The value of any respective element shall be greater than 0 and less than or equal to 100,000.

## Output Format

Print 3 lines of output in the following order:

1. Print the *mean* on a new line, to a scale of 1 decimal place (i.e., 12.3, 7.0).
2. Print the *median* on a new line, to a scale of 1 decimal place (i.e., 12.3, 7.0).
3. Print the *mode* on a new line; if more than one such value exists, print the numerically smallest one.

```python
# read in the first input
# (str) --> int

num_elements = int(input()) # supposed to be equal to len(data) as below

# read in the second input
# (str of N space-seperated ints) --> list of ints

data = list(map(int, input().split()))


#------------------
# calulate the mean
#------------------

# Sum total of the elements in the data list divided by the number of elements

# Assigned value to num_elems rather than using len(data) directly, just in case 
# hackerrank was going to test input_line_1 value against constraint parameters. 

total = 0

for item in data:
    total = total + item
    
mean = (total / num_elems) # alternately mean = (total / len(data))

#--------------------
# calulate the median
#--------------------

# sort the data list
data.sort()

# determine even or odd
even = True
if len(data)%2 == 1:
    even = False


if even:
    slice_start = (len(data)//2) - 1
    slice_end   = (len(data)//2) + 1
    median      = sum(data[slice_start:slice_end]) / 2
else:
    median      = data[len(data)//2]

#--------------------
# count the frequency of each number in the data into a dictionary
#--------------------
numbers = dict()

for item in data:
    if item not in numbers:
        numbers[item] = 1
    else:
        numbers[item] = numbers[item] + 1

        
    
#--------------------
# calulate the mode
#--------------------

max_count = 0

for key in numbers:
    if numbers[key] > max_count:
        max_count = numbers[key]

max_modal_elements = []

for key in numbers:
    if numbers[key] == max_count:
        max_modal_elements.append(key)
        
mode = min(max_modal_elements)




#--------------------
# print the output
#--------------------

print("%.1f" % mean)
print("%.1f" % median)
print(mode)
```
